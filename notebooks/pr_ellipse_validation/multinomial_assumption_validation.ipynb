### Generate data

```python
def generate_classifier(n_train_samples, n_test_samples, n_holdout_sets):
    t_samples = n_train_samples + n_test_samples * (n_holdout_sets + 1)
    # generate 2 class dataset
    X, y = make_classification(n_samples=t_samples, n_classes=2, n_redundant=2, n_features=4, random_state=23445342)
    sidx = np.random.choice(np.arange(y.size), replace=False, size=t_samples)
    X = X[sidx, :]
    y = y[sidx]

    # split into train/test/holdout sets
    ub_test = n_train_samples + n_test_samples
    X_train = X[:n_train_samples, :]
    X_test = X[n_train_samples:ub_test, :]
    X_holdout = X[ub_test:, :]
    
    y_train = y[:n_train_samples]
    y_test = y[n_train_samples:ub_test]
    y_holdout = y[ub_test:].reshape(n_test_samples, n_holdout_sets, order='F')
    
    # fit a model
    model = LogisticRegression(solver='lbfgs', penalty='none')
    
    model.fit(X_train, y_train)
    # predict probabilities
    y_test_score = model.predict_proba(X_test)[:, 1]
    test_conf_mat = mmu.confusion_matrix(y_test, score=y_test_score, threshold=0.5,).flatten()
    y_holdout_score = model.predict_proba(X_holdout)[:, 1].reshape(n_test_samples, n_holdout_sets, order='F')
    ho_conf_mats = mmu.confusion_matrices(y=y_holdout, score=y_holdout_score, threshold=0.5)
    true_conf_mat = mmu.confusion_matrix(y=y_holdout.ravel(), score=y_holdout_score.ravel(), threshold=0.5).flatten()
    return (
        true_conf_mat,
        test_conf_mat,
        ho_conf_mats
    )
```

```python
test_sizes = np.floor(np.geomspace(100, 2000, 20)).astype(np.int64)
N = test_sizes.size
```

```python
from mmu.methods import precision_recall_uncertainty
```

```python
def prec_sigma(cm):
    cm = cm.flatten()
    return np.sqrt((cm[3] * cm[1]) / np.power((cm[3] + cm[1]).astype(np.float64), 3))

def rec_sigma(cm):
    cm = cm.flatten()
    return np.sqrt((cm[3] * cm[2]) / np.power((cm[3] + cm[1]).astype(np.float64), 3))
```

```python
ho_sigmas = np.empty((N, 2))
true_sigmas = np.empty((N, 2))
test_sigmas = np.empty((N, 2))
test_conf_mats = np.empty((N, 4))

for i, test_size in enumerate(test_sizes):
    print(f'test_size: {test_size}')
    true_conf_mat, test_conf_mat, ho_conf_mats = generate_classifier(test_size, test_size, 1000)
    ho_mtr = mmu.binary_metrics_confusion_matrices(ho_conf_mats)[:, [1, 3]]
    ho_sigmas[i, :] = np.std(ho_mtr, axis=0, ddof=1)
    true_sigmas[i, 0] = prec_sigma(true_conf_mat)
    true_sigmas[i, 1] = rec_sigma(true_conf_mat)
    test_sigmas[i, 0] = prec_sigma(test_conf_mat)
    test_sigmas[i, 1] = rec_sigma(test_conf_mat)
```
